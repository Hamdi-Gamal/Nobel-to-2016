```python

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("/kaggle/input/olympics-124-years-datasettill-2020/Athletes_summer_games.csv")
```

```python
df.tail(20)
```

```python
plt.figure(figsize=(9,4))
plt.title("Number Of Medals Over The Year")
df.Year.value_counts().sort_index().plot()
```

```python
itemnames = df["Team"].value_counts().index[:10]
itemvalue = df["Team"].value_counts().values[:10]
plt.figure(figsize=(20, 9))
plt.ylabel("Count", fontsize="medium")
plt.xlabel("Team", fontsize="medium")
plt.title("Top 10 Teams")
plt.bar(itemnames, itemvalue, width =0.7, color ="purple", linewidth=0.4)
for i in range (len(itemnames)):
    plt.text(i, itemvalue[i],itemvalue[i],ha= "center", va="bottom")
plt.show()
```

```python
itemnames = df["City"].value_counts().index[:10]
itemvalue = df["City"].value_counts().values[:10]
plt.figure(figsize=(20, 9))
plt.ylabel("Count", fontsize="medium")
plt.xlabel("City", fontsize="medium")
plt.title("Top 10 City")
plt.bar(itemnames, itemvalue, width =0.7, color ="purple", linewidth=0.4)
for i in range (len(itemnames)):
    plt.text(i, itemvalue[i],itemvalue[i],ha= "center", va="bottom")
plt.show()
```

```python
labels= df.Sex.value_counts().index
colors = ['royalblue', 'orangered']
gender= df.Sex.value_counts().values
plt.figure(figsize = (5,5))
plt.pie(gender, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('proportion of genders', color = 'black', fontsize = 15)


```
